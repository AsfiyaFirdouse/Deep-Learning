```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/drive/MyDrive/Social Media Usage and Emotional Well-Being.csv")
df
```

```python
pd.to_numeric(df['Age'], downcast='integer')
df
```

```python
df=df.drop('User_ID',axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col=['Gender','Platform','Dominant_Emotion']
for col_name in col:
  df[col_name]=le.fit_transform(df[col_name])
df
```

```python
df.info()
```

```python
X = df.drop('Dominant_Emotion',axis=1)
y = df['Dominant_Emotion']
y=np.array(df['Dominant_Emotion'])

from tensorflow.keras.utils import to_categorical

y=to_categorical(y)
y
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x_scale= min_max_scaler.fit_transform(X)
print(x_scale)
```

```python
# Splitting training and test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.20,random_state=0)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import LeakyReLU
from keras.layers import Dense
from tensorflow.keras.layers import Dropout

model=Sequential()
model.add(Dense(50, activation='relu', input_shape=(8,)))
model.add(Dense(42, activation='relu'))
model.add(Dropout(0.12))
model.add(Dense(30, activation=LeakyReLU(alpha=0.01)))
#output layer
model.add(Dense(6, activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=10, epochs=250)
```

```python
y_pred=model.predict(X_test)
y_pred
y_pred = np.argmax(y_pred, axis=1) + 1
y_pred
np.column_stack((y_pred, y_test))
```

```python
from sklearn.metrics import accuracy_score

loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```
